## Висновки за результатами A/B тестування оновленого екрана підписки

**Дата:** 2 квітня 2025 року

**Контекст:**

Проведено A/B тестування з метою оцінки впливу оновленого дизайну екрана пропозиції тижневої підписки вартістю $4.99 на конверсію користувачів. У початковому дизайні (група A) користувачу просто пропонувалася підписка. В альтернативному дизайні (група B) користувачу пропонувалася та ж підписка за тією ж ціною, але з позначенням "знижка 50%". Ціль тесту полягала у визначенні, чи призведе акцентування на знижці до збільшення кількості покупок підписки.

**Період тестування:** 2023-07-03 - 2023-07-25 (22 дні)

**Розподіл користувачів:**

* Група A (контрольна): 10013 користувачів
* Група B (тестова, з позначенням знижки): 9985 користувачів

**Результати:**

| Метрика             | Група A (Первинний дизайн) | Група B (Новий дизайн) |
| ------------------- | --------------------------- | ---------------------- |
| Кількість користувачів | 10013                       | 9985                   |
| Кількість конверсій  | 611                         | 889                    |
| Рівень конверсії    | 6.10%                       | 8.90%                  |

**Статистичний аналіз:**

Для оцінки статистичної значущості різниці в рівнях конверсії між групами було використано критерій хі-квадрат.

* **Статистика хі-квадрат:** 56.14
* **P-значення:** 0.000

Отримане p-значення (0.000) є меншим за встановлений рівень значущості (α = 0.05). Це дає підстави відхилити нульову гіпотезу про відсутність різниці між рівнями конверсії в групах A та B.

**Висновки:**

Результати A/B тестування показали статистично значуще збільшення рівня конверсії на екрані підписки після впровадження нового дизайну з позначенням "знижка 50%".

* Рівень конверсії в групі B (8.90%) виявився на **2.80%** вищим, ніж у групі A (6.10%).
* Статистичний аналіз підтверджує, що ця різниця не є випадковою.

**Рішення:**

На основі отриманих статистично значущих результатів, ми рекомендуємо **повністю впровадити новий дизайн екрана підписки (група B)** для всіх користувачів. Акцентування на знижці позитивно впливає на рішення користувачів щодо придбання тижневої підписки.

**Подальші кроки:**

* Продовжити моніторинг рівня конверсії після повного впровадження нового дизайну.
* Розглянути можливість проведення подальших A/B тестів для оптимізації інших елементів екрана підписки або інших екранів онбордингу.
* Проаналізувати вплив зміни на інші ключові метрики, такі як утримання користувачів, які придбали підписку.
